```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python

df.info()

# Summary statistics
df.describe()

# Missing values
df.isnull().sum()
```

```python
df.hist(figsize=(12,8), bins=30)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(df["Age"])
plt.title("Age Boxplot")
plt.show()

sns.boxplot(df["Fare"])
plt.title("Fare Boxplot")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(), 
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Sex")
plt.show()

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Class")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Sex")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.kdeplot(df[df["Survived"]==1]["Age"], shade=True, label="Survived")
sns.kdeplot(df[df["Survived"]==0]["Age"], shade=True, label="Not Survived")
plt.title("Age Distribution by Survival")
plt.legend()
plt.show()
```

```python
sns.kdeplot(df[df["Survived"]==1]["Fare"], shade=True, label="Survived")
sns.kdeplot(df[df["Survived"]==0]["Fare"], shade=True, label="Not Survived")
plt.title("Fare Distribution by Survival")
plt.legend()
plt.show()
```
